```python
# data ready from 2016-12-02
from datetime import datetime, timedelta, date
import requests
import json
import pandas as pd

# start_date = datetime.date(2016, 12, 2)
# end_date = datetime.today()


'''
# The size of each step in days
day_delta = datetime.timedelta(days=1) # <class 'datetime.timedelta'>

start_date_1 = datetime.date.today() # <class 'datetime.datetime'>
# datetime.today().date() ------  <class 'datetime.date'>
# date(2021, 8, 4) ------  <class 'datetime.date'>
end_date_1 = start_date + day_delta # <class 'datetime.date'>
'''
def daterange(start_date, end_date):
    # The size of each step in days
    day_delta = timedelta(days=1) # <class 'datetime.timedelta'>
    end_date_adjusted = end_date + day_delta # <class 'datetime.date'>
    for n in range(int ((end_date_adjusted - start_date).days)):
        yield start_date + timedelta(n)

start_date = date(2016, 12, 2) # <class 'datetime.date'>
end_date = datetime.today().date() # <class 'datetime.date'>

dates_list = []
for single_date in daterange(start_date, end_date):
    dates_list.append(single_date.strftime("%Y-%m-%d"))
df = pd.DataFrame(dates_list)
df.to_csv('dates.csv', index=False)
```
