```python
import pandas as pd
import numpy as np
import json
import os
```

```python
path = 'file_json/'

for filename in os.listdir(path):
    f = os.path.join(path, filename)
    if os.path.isfile(f):
        print(f)
        # opening each json file
        data = json.load(open(f))
        df = pd.DataFrame(data['metadata']['stations'])
        #df.describe()
        readings = data['items']

        if readings != []:

            # getting list of timestamp for each json file
            timestamp = []
            for i in range(len(readings)):
                if readings[i]['timestamp'] not in timestamp:
                    timestamp.append(readings[i]['timestamp'])

            # getting list of readings for each json file
            list_of_value_data = []
            list_of_station_data = []

            for x in range(len(readings)):
                raw_data = readings[x]['readings']
                #print(raw_data)
                value_data = []
                station_data = []
                for y in range(len(raw_data)):
                    try:
                        value_data.append(raw_data[y]['value'])
                        station_data.append(raw_data[y]['station_id'])
                    except KeyError:
                        continue

                list_of_value_data.append(value_data)
                list_of_station_data.append(station_data)


            # getting max list of station_id/value data as dataframe index
            len_readings_1_timestamp = []
            for a in list_of_value_data:
                len_readings_1_timestamp.append(len(a))

            max_len = max(len_readings_1_timestamp)
            max_index = len_readings_1_timestamp.index(max_len) 

            # creating initial dataframe
            new_df = pd.DataFrame({timestamp[0]:list_of_value_data[max_index]}, columns = timestamp, index = list_of_station_data[max_index])
            #new_df.head()

            # remove data of max index (already inside dataframe)
            list_of_station_data.pop(max_index)
            list_of_value_data.pop(max_index)
            timestamp.pop(0)

        
            # populating the dataframe with readings
            if (len(timestamp) == len(list_of_station_data) == len(list_of_value_data)):
                for b in range(len(list_of_station_data)):
                    for c in range(len(list_of_station_data[b])):
                        station_id = list_of_station_data[b][c]
                        if station_id in new_df.index:
                            new_df.loc[station_id, timestamp[b]] = list_of_value_data[b][c]
                        else:
                            continue
            else:
                print('length of data is not the same, dataframe cannot be populated')
                break


            # saving dataframe to csv
            new_path = 'clean_csv_data/' + filename.split('.')[0] + '.csv'
            new_df.to_csv(new_path, encoding='utf-8')
        
        else:
            print('no readings in json file')

    else:
        print('path indicated is not a file')
        break
```
