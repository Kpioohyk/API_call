```python
import pandas as pd
import numpy as np
import json

filename = '2017-03-05.json'
```

### Opening json file

```python
data2 = json.load(open(filename))

df = pd.DataFrame(data2['metadata'])
```

```python
readings = data2["items"]
readings
```

```python
timestamp = []
for i in range(len(readings)):
    if readings[i]['timestamp'] not in timestamp:
        timestamp.append(readings[i]['timestamp'])
        
```

```python
timestamp[:5]
```

```python
list_of_value_data = [] 
list_of_station_data = []

for i in range(len(readings)):
    raw_data = (readings[i]['readings']) 
    #print(raw_data)
    value_data = []
    station_data = []
    for j in range(len(raw_data)): 
        #print(raw_data[j]['value'])
        value_data.append(raw_data[j]['value'])
        station_data.append(raw_data[j]['station_id'])
        
    list_of_value_data.append(value_data)
    list_of_station_data.append(station_data)
    
```

```python
len_readings_1_timestamp = []
for i in list_of_value_data:
    len_readings_1_timestamp.append(len(i))
```

```python
max_len = max(len_readings_1_timestamp)

max_index = len_readings_1_timestamp.index(max_len)
max_index
```
